# Fill missing engagement rates with the average engagement rate

```python
df['60_day_eng_rate'].fillna(df['60_day_eng_rate'].mean(), inplace=True)
```

# Drop Rows With Any Missing Values

```python
df.dropna(inplace=True)
df
```

# Drop any duplicate rows if present

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.duplicated().sum()
```

# Handle missing values

```python
df.isnull().sum()
```

```python
mode_country = df['country'].mode()[0]
df['country'] = df['country'].fillna(mode_country)
```

# Exploratory Data Analysis

```python
print(df[['influence_score']].describe())
```

```python
print(df[['followers',
          'avg_likes','60_day_eng_rate','new_post_avg_like']].describe())
```

```python
country= df['country'].value_counts()[:20].to_list()
name_countries = df['country'].value_counts().index[:20].to_list()

name_countries.append("Others")
max20 = sum(country)
others = len(df) - max20

country.append(others)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=country, y=name_countries, palette='viridis')
plt.title('Distribution of Influencers by Country')
plt.xlabel('Number of Influencers')
plt.ylabel('Country')
plt.show()
```

# Relationship B/W Followers and Engagement Rate
