# Visualizations

# Distribution of Ratings

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=20, kde=True)

plt.title('Distribution of App Ratings')
plt.show()
```

```python
# Count of Apps by Category
plt.figure(figsize=(12, 8))
sns.countplot(y='Category', data=df,
order=df['Category'].value_counts().index)
plt.title('Count of Apps by Category')
plt.show()
```

# Relationship between Installs and Rating

```python
plt.figure(figsize=(8,6))
sns.histplot(x='Rating',y='Installs',hue='Category',data=df)
plt.title("Relationship between Installs & Ratings")
plt.show()
```

# Analyzing key metrics

# Average rating by category

```python
avg_rating_by_category = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print(avg_rating_by_category)
```

# Most Popular apps by installs

```python
most_installed_apps = df[['App','Installs']].sort_values(by='Installs',ascending = False).head(10)
print(most_installed_apps)
```

```python
top_genres = df['Genres'].value_counts().head(5)
print(top_genres)
```

```python
df['Category'].value_counts().head(10)
```

# Average Rating by Category

```python
avg_rating = df.groupby('Category')['Rating'].mean().reset_index()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_rating['Category'], avg_rating['Rating'], color='purple')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Average Rating by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
